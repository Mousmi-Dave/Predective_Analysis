#Modelling


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train=data.drop(['price','id','date'],axis=1)
test=data['price']
```

```python
X_train,X_test, y_train,y_test=train_test_split(train,test,test_size=0.3,random_state=1)
```

```python
regr=LinearRegression() #An instance of the LinearRegression class is created and assigned to the variable regr.
```

```python
regr.fit(X_train,y_train)
```

```python
pred=regr.predict(X_test)
pred
```

```python
regr.score(X_test,y_test)
```
